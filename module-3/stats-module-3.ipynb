```python
import pandas as pd
import networkx as nx
phases = {}
G = {}
for i in range(1,12): 
  var_name = "phase" + str(i)
  file_name = "https://raw.githubusercontent.com/ragini30/Networks-Homework/main/" + var_name + ".csv"
  phases[i] = pd.read_csv(file_name, index_col = ["players"])
  phases[i].columns = "n" + phases[i].columns
  phases[i].index = phases[i].columns
  phases[i][phases[i] > 0] = 1
  G[i] = nx.from_pandas_adjacency(phases[i])
  G[i].name = var_name
```

```python
B = {}
meanB = {}
for i in range(1,12):
  B[i] = nx.betweenness_centrality(G[i], normalized = True)

for i in range(1,12):
  for p in B[i]:
    if p not in meanB:
      meanB[p] = 0
    else:
      meanB[p] += B[i][p]

print(meanB)

import operator
sorted_d = dict( sorted(meanB.items(), key=operator.itemgetter(1),reverse=True))
print(sorted_d)
```
